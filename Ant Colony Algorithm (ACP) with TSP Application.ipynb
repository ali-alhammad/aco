## <span style='color:blue;'><u>What is Ant Colony Algorithm?</u></span>
Ant Colony Optimization is a __probabilistic__ technique for solving computational problems which can be reduced to finding good paths through graphs.  Artificial ants stand for multi-agent methods inspired by the behavior of real ants.

### <span style='color:blue;'> Introduction 
- Ant Colony Optimization **(ACO)** was introduced in 1992 by the Italian scientist Marco Dorigo.
- The first version of the ACO was named Ant systems
- The algorithm was inspired by the way ants in a colony cooperate to find the shortest path when searching for food 


### <span style='color:blue;'>Stigmergy</span>
- Stigmergy is a mechanism of indirect coordination, through the environment, between agents or actions. The principle is that the trace left in the environment by an individual agent (or action) stimulates the performance of a succeeding agent (or action) by the same or different agent.
- Species in nature build a complex decenteralized intelligence systems by performing specific action without any planning, coordination, or direct communication.
- Search for food by ants is an optimization process where ants try to find the maximum amount of food by consuming the minimum amount of energy. 
- This can be achieved by finding the shortest path form the colony to any food source


### <span style='color:blue;'>How Ants Communicate Indirectly</span>
- Ants leave a trace in a form chemicals called pheromone when they travel on a path

- Other ants follow the traces of the pheromone (if one side of the path there are more pheromone, then the ants will follow that path rather than the other path)

- This enables the ants to figure out the shortest path

- The path with high and intense pheromone level is considered the shortest because more trips are taken along this path (ants produce many pheromones  for different purposes and one of them is used to mark the paths) .  
- This means ants make decision based on probabilities 

### <span style='color:blue;'>1.4 ACO Explained</span>

Solutions are found through an iterative process. In each iteration, several ants are allowed to find a solution that “visits” every node of the world. The amount of pheromone on each edge is updated according to the length of the solutions in which it was used. The ant that traveled the least distance is considered to be the local best solution. If the local solution has a shorter distance than the best from any previous iteration, it then becomes the global best solution. The elite ant(s) then deposit their pheromone along the path of the global best solution to strengthen it further, and the process repeats.

### <span style='color:blue;'>Full Implementation of the ACO</span>

```python
import numpy as np
from numpy import inf

#given values for the problems

d = np.array([[0 , 10, 12, 11, 14]
             ,[10 ,0 , 13, 15, 8 ]
             ,[12, 13, 0 , 9 , 14]
             ,[11, 15, 9 , 0 , 16]
             ,[14, 8 , 14, 16, 0 ]])


iteration = 100
n_ants = 5
n_citys = 5
```

```python
# intialization part

m = n_ants
n = n_citys
e = .5         #evaporation rate
alpha = 1     #pheromone factor
beta = 2       #visibility factor
```

```python
#calculating the visibility of the next city visibility(i,j)=1/d(i,j)

visibility = 1/d
visibility[visibility == inf ] = 0
```

```python
#intializing pheromne present at the paths to the cities

pheromne = .1*np.ones((m,n))

#intializing the rute of the ants with size rute(n_ants,n_citys+1) 
#note adding 1 because we want to come back to the source city

rute = np.ones((m,n+1), dtype=int)
```

```python
for ite in range(iteration):
    
    rute[:,0] = 1          #initial starting and ending positon of every ants '1' i.e city '1'
    
    for i in range(m):
        
        temp_visibility = np.array(visibility)         #creating a copy of visibility
        
        for j in range(n-1):
            #print(rute)
            
            combine_feature = np.zeros(5)     #intializing combine_feature array to zero
            cum_prob = np.zeros(5)            #intializing cummulative probability array to zeros
            
            cur_loc = int(rute[i,j]-1)        #current city of the ant
            
            temp_visibility[:,cur_loc] = 0     #making visibility of the current city as zero
            
            p_feature = np.power(pheromne[cur_loc,:],beta)         #calculating pheromne feature 
            v_feature = np.power(temp_visibility[cur_loc,:],alpha)  #calculating visibility feature
            
            p_feature = p_feature[:,np.newaxis]                     #adding axis to make a size[5,1]
            v_feature = v_feature[:,np.newaxis]                     #adding axis to make a size[5,1]
            
            combine_feature = np.multiply(p_feature,v_feature)     #calculating the combine feature
                        
            total = np.sum(combine_feature)                        #sum of all the feature
            
            probs = combine_feature/total   #finding probability of element probs(i) = comine_feature(i)/total
            
            cum_prob = np.cumsum(probs)     #calculating cummulative sum
            #print(cum_prob)
            r = np.random.random_sample()   #randon no in [0,1)
            #print(r)
            city = int(np.nonzero(cum_prob>r)[0][0]+1)       #finding the next city having probability higher then random(r) 
            #print(city)
            
            rute[i,j+1] = city              #adding city to route 
           
        left = list(set([i for i in range(1,n+1)])-set(rute[i,:-2]))[0]     #finding the last untraversed city to route
        
        rute[i,-2] = int(left)                   #adding untraversed city to route
       
    rute_opt = np.array(rute)               #intializing optimal route
    
    dist_cost = np.zeros((m,1))             #intializing total_distance_of_tour with zero 
    
    for i in range(m):
        
        s = 0
        for j in range(n-1):
            
            s = s + d[int(rute_opt[i,j])-1,int(rute_opt[i,j+1])-1]   #calcualting total tour distance
        
        dist_cost[i]=s                      #storing distance of tour for 'i'th ant at location 'i' 
       
    dist_min_loc = np.argmin(dist_cost)             #finding location of minimum of dist_cost
    dist_min_cost = dist_cost[dist_min_loc]         #finging min of dist_cost
    
    best_route = rute[dist_min_loc,:]               #intializing current traversed as best route
    pheromne = (1-e)*pheromne                       #evaporation of pheromne with (1-e)
    
    for i in range(m):
        for j in range(n-1):
            dt = 1/dist_cost[i]
            pheromne[int(rute_opt[i,j])-1,int(rute_opt[i,j+1])-1] = pheromne[int(rute_opt[i,j])-1,int(rute_opt[i,j+1])-1] + dt   
            #updating the pheromne with delta_distance
            #delta_distance will be more with min_dist i.e adding more weight to that route  peromne

print('route of all the ants at the end :')
print(rute_opt)
print()
print('best path :', best_route)
print('cost of the best path',int(dist_min_cost[0]) + d[int(best_route[-2])-1,0])
   
```

To use ACOpy in a project, you simply use it to create a Solver and a Colony:


```python
# Solvers control the parameters related to pheromone deposit and evaporation.
# rho : the percentage of pheromone to evaporate
# q : the amount of pheromone
solver = solver = acopy.Solver(rho=0.3, q=1)

# alpha:  relative factor for edge pheromone
#  beta: relative factor for edge weight (distance)
colony = acopy.Colony( alpha=1, beta=1)
```

We can use the acopy solver and the colony to solve any weighted networkx graph. 


```python
#solver.optimize(G,colony)

# limit:  maximum number of iterations to perform (default is unlimited so it will run forever)
tour = solver.solve(G, colony,
                    gen_size = 20, # gen_size : number of Ant s to use (default is one per graph node)
                    limit=100)
```

```python

def draw_map(m, scale=0.2):
    # draw a shaded-relief image
    m.shadedrelief(scale=scale)
    
    # lats and longs are returned as a dictionary
    lats = m.drawparallels(np.linspace(-90, 90, 13))
    lons = m.drawmeridians(np.linspace(-180, 180, 13))

    # keys contain the plt.Line2D instances
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)
    
    # cycle through these lines and set the desired style
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')
```

## <span style='color:blue;'><u>When  ACO is appropriate? and when it is not? and why?</u></span>


### <span style='color:blue;'>When to Use ACO? Why?</span>
ACO is most suitable to solve combinatorial optimization problems that can be represented using graph. 

I am listing some of the most popular applications with the algorithm names.

<img src ='https://github.com/ali-alhammad/aco/blob/main/Current-applications-of-ACO-algorithms-Applications-are-listed-by-class-of-problems-and.png?raw=true' />

__Source__: Dorigo, Marco & Di Caro, Gianni & Gambardella, Luca Maria. (1999). Ant Algorithms for Discrete Optimization.” Artificial Life 5, 137-172. Artificial Life. 5. 137-172. 10.1162/106454699568728. 

<span style="color:green;">They have an advantage over simulated annealing and genetic algorithm approaches of similar problems when the graph may change dynamically; the ant colony algorithm can be run continuously and adapt to changes in real time.</span>

### <span style='color:blue;'>When not to use ACO? Why?</span>
The main issues concern with the ACO algorithm are as follows: 
- The convergence speed of the algorithm changes rapidly as number of iterations goes high. 
- If the problem requires a deterministic solution, then there is no guarantee that ACO will provide the same solution every time since ACO is probabilistic and stochastic in nature.
- Convergence is guaranteed, but time to convergence uncertain
- Coding is not straightforward
- Sometimes ACO trapped easily into stagnation phase where no further  modification can be done. This happens mostly for the few number of cities and larger number of ants.

```python
import acopy
import networkx as nx
import matplotlib.pyplot as plt
from itertools import chain
from mpl_toolkits.basemap import Basemap
import folium
from folium.features import DivIcon
from folium import plugins
import tsplib95
```

```python
import tsplib95
#original_url = 'https://dev.heuristiclab.com/trac.fcgi/export/7465/branches/GeneralizedQAP/HeuristicLab.Problems.Instances.TSPLIB/3.3/Data/TSP/bayg29.tsp'
#source_url = 'http://raw.githubusercontent.com/ali-alhammad/aco/7fa3100d5b707569a4c072ec388642ce49a50c54/bayg29.tsp'

problem = tsplib95.load('bayg29.tsp')
#problem = tsplib95.load(source_url)
G2 = problem.get_graph()
```

Solving is easy. Let’s do 100 iterations with a default number of ants:

```python
# define a colony

# alpha:  relative factor for edge pheromone
#  beta: relative factor for edge weight (distance)
colony2 = acopy.Colony( alpha=1, beta=1)
```

```python
# Define a solver
# Solvers control the parameters related to pheromone deposit and evaporation.
# rho : the percentage of pheromone to evaporate
# q : the amount of pheromone
solver2 = solver = acopy.Solver(rho=0.3, q=1)
```

```python
# solve to find the path
tour2 = solver2.solve(G2, colony2, limit=10)
```

How good was the best tour found? Let us look:

```python
print("The lenght of the traveled path in Kilometers: ")
print(tour2.cost)
```

You can list the solution tour in terms of the nodes or edges:

```python
print("The cities in the oder they were visited:")
print(tour2.nodes)
```

```python
print("The path that was followed (from_city , to_city)")
print(tour2.path)
```
